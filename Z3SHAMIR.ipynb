```python
# Secret and coefficients
S = 1954
a1 = 43
a2 = 12

# Prime field size, choose a large enough prime number
p = 104729  # Example prime, should be larger than the coefficients and secret

# Function to compute the share given x
def compute_share(x, S, a1, a2, p):
    y = (S + a1 * x + a2 * x**2) % p
    return (x, y)

# Generate 5 shares
shares = [compute_share(x, S, a1, a2, p) for x in range(1, 6)]

# Print the shares
for x, y in shares:
    print(f"Share {x}: ({x}, {y})")
```

```python
# Given shares
shares = [(1, 2009), (2, 2088), (3, 2191), (4, 2318), (5, 2469)]

# Prime field size
p = 104729  # Example prime number, should be larger than the secret and coefficients

# Define the ranges for brute force
secret_range = range(1900, 2001)
coeff1_range = range(1, 101)
coeff2_range = range(1, 101)

# Function to compute the polynomial value
def polynomial(x, S, a1, a2, p):
    return (S + a1 * x + a2 * x**2) % p

# Brute force search
found = False
for S in secret_range:
    for a1 in coeff1_range:
        for a2 in coeff2_range:
            valid = True
            for x, y in shares:
                if polynomial(x, S, a1, a2, p) != y:
                    valid = False
                    break
            if valid:
                print(f"Secret: {S}")
                print(f"Coefficient 1: {a1}")
                print(f"Coefficient 2: {a2}")
                print(f"Shares are valid with these parameters.")
                found = True
                break
        if found:
            break
    if found:
        break

if not found:
    print("No valid secret and coefficients found in the given ranges.")
```
