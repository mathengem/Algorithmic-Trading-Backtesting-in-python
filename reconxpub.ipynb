```python
import numpy as np
from scipy.optimize import brute

def polynomial(x, a, b, c):
    return a + b * x + c * x**2

# Define the objective function to minimize
def objective(params):
    # Unpack the parameters
    a, b, c = params

    # Calculate the sum of squared differences between the polynomial and the given points
    diff1 = polynomial(x1, a, b, c) - y1
    diff2 = polynomial(x2, a, b, c) - y2
    # Add a penalty for values outside the specified range
    penalty = sum(np.maximum(y_min - polynomial(x, a, b, c), 0) for x in range(1, 5))
    return diff1**2 + diff2**2 + penalty

# Define the range for each parameter as a tuple of tuples
ranges = ((1900, 1999), (1, 99), (1, 99))

# Perform the brute force optimization
result = brute(objective, ranges, finish=None)

# Extract the optimal coefficients
a_opt, b_opt, c_opt = result

# Evaluate the polynomial at the specified x values
for x in range(1, 5):
    y = polynomial(x, a_opt, b_opt, c_opt)
    print(f"Share for x={x}: ({x}, {y})")
```

```python
import base58
from bip32 import BIP32

# Function to convert zpub to xpub
def convert_zpub_to_xpub(zpub):
    # Decode the zpub key
    decoded = base58.b58decode_check(zpub)
    # Change the version bytes to xpub (0x0488B21E)
    xpub_bytes = b'\x04\x88\xb2\x1e' + decoded[4:]
    xpub = base58.b58encode_check(xpub_bytes).decode()
    return xpub

# Extended Public Key (zpub)
zpub = "zpub6qdEDkv51FpxX6g1rpFGckmiL46vV8ccmtEgPAkj3qj8N4ZZHyXDRA9RwpTiFK2Kb8vRaDmSmwgX6rfB4t2K8Ktdq8ExQ6fumKpn2ndJCqL"

# Convert zpub to xpub
try:
    xpub = convert_zpub_to_xpub(zpub)
    print(f"Converted xpub: {xpub}")

    # Initialize BIP32 object with the extended public key
    bip32 = BIP32.from_xpub(xpub)

    # Derive child keys
    # Example derivation path: m/0/0
    child_index = 0
    child_path = f"m/0/{child_index}"

    # Derive child public key
    child_pub_key = bip32.get_pubkey_from_path(child_path).hex()
    print(f"Child Public Key ({child_path}): {child_pub_key}")

    # Derive multiple child keys
    for i in range(5):
        child_path = f"m/0/{i}"
        child_pub_key = bip32.get_pubkey_from_path(child_path).hex()
        print(f"Child Public Key ({child_path}): {child_pub_key}")
except Exception as e:
    print(f"Error: {e}")
```
