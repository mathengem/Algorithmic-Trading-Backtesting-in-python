```python
import os
import secrets
import numpy as np
import pycuda.autoinit
import pycuda.driver as cuda
from pycuda.compiler import SourceModule
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives import serialization, hashes
from cryptography.hazmat.primitives.asymmetric import padding

# Function to generate a secure entropy source
def generate_entropy(length=32):
    return secrets.token_bytes(length)

# Function to generate RSA key pair
def generate_rsa_key_pair():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048,
    )
    public_key = private_key.public_key()
    return private_key, public_key

# Function to serialize keys to PEM format
def serialize_keys(private_key, public_key):
    private_pem = private_key.private_bytes(
        encoding=serialization.Encoding.PEM,
        format=serialization.PrivateFormat.TraditionalOpenSSL,
        encryption_algorithm=serialization.NoEncryption()
    )
    public_pem = public_key.public_bytes(
        encoding=serialization.Encoding.PEM,
        format=serialization.PublicFormat.SubjectPublicKeyInfo
    )
    return private_pem, public_pem

# Function to generate an address (for demonstration, we'll use the public key hash)
def generate_address(public_key):
    public_bytes = public_key.public_bytes(
        encoding=serialization.Encoding.DER,
        format=serialization.PublicFormat.SubjectPublicKeyInfo
    )
    address = hashes.Hash(hashes.SHA256())
    address.update(public_bytes)
    return address.finalize().hex()

# CUDA kernel for vanity pattern matching
mod = SourceModule("""
__global__ void match_vanity_pattern(char *addresses, char *pattern, int *results, int num_addresses, int pattern_length) {
    int idx = blockIdx.x * blockDim.x + threadIdx.x;
    if (idx < num_addresses) {
        int match = 1;
        for (int i = 0; i < pattern_length; i++) {
            if (addresses[idx * 64 + i] != pattern[i]) {
                match = 0;
                break;
            }
        }
        results[idx] = match;
    }
}
""")

# Function to check if the address matches a vanity pattern using GPU
def matches_vanity_pattern_gpu(addresses, pattern):
    num_addresses = len(addresses)
    pattern_length = len(pattern)

    # Convert addresses to a single bytes-like object
    addresses_bytes = ''.join(addresses).encode('utf-8')

    # Allocate memory on the GPU
    addresses_gpu = cuda.mem_alloc(len(addresses_bytes))
    pattern_gpu = cuda.mem_alloc(pattern_length)
    results_gpu = cuda.mem_alloc(num_addresses * 4)

    # Copy data to the GPU
    cuda.memcpy_htod(addresses_gpu, addresses_bytes)
    cuda.memcpy_htod(pattern_gpu, pattern.encode('utf-8'))

    # Launch the kernel
    block_size = 256
    grid_size = (num_addresses + block_size - 1) // block_size
    match_vanity_pattern = mod.get_function("match_vanity_pattern")
    match_vanity_pattern(addresses_gpu, pattern_gpu, results_gpu, np.int32(num_addresses), np.int32(pattern_length), block=(block_size, 1, 1), grid=(grid_size, 1))

    # Copy results back to the host
    results = np.empty(num_addresses, dtype=np.int32)
    cuda.memcpy_dtoh(results, results_gpu)

    return results

# Main function to generate vanity address
def generate_vanity_address(pattern):
    addresses = []
    while True:
        entropy = generate_entropy()
        private_key, public_key = generate_rsa_key_pair()
        address = generate_address(public_key)
        addresses.append(address)

        if len(addresses) >= 1000:  # Check in batches of 1000
            results = matches_vanity_pattern_gpu(addresses, pattern)
            for i, result in enumerate(results):
                if result == 1:
                    private_pem, public_pem = serialize_keys(private_key, public_key)
                    return private_pem, public_pem, addresses[i]
            addresses = []

# Example usage
if __name__ == "__main__":
    try:
        pattern = "1ucXXZQSEf4z"  # Example vanity pattern
        private_pem, public_pem, address = generate_vanity_address(pattern)

        # Print the keys and address in a readable format
        print("Private Key:\n")
        print(private_pem.decode())
        print("\nPublic Key:\n")
        print(public_pem.decode())
        print("\nAddress:\n")
        print(address)
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
def lagrange_interpolation(x, x_s, y_s, p):
    """
    Lagrange Interpolation to find the polynomial at a given x.
    """
    k = len(x_s)
    assert k == len(y_s), "x and y must be of the same length"

    def _basis(j):
        b = 1
        for m in range(k):
            if m != j:
                b *= (x - x_s[m]) * pow(x_s[j] - x_s[m], p-2, p)
                b %= p
        return b

    result = 0
    for j in range(k):
        result += y_s[j] * _basis(j)
        result %= p
    return result

# Given shares
shares = [
    (1, "1ed2cc1d323cb527df6327c23c1684e14396c3d8252a045c3699f3556d01653f"),
    (2, "136444b1b9b3505a9e246c2b122a2de29311a99a00b4490b105550528676b998")
]

# Prime number p
p = 2**256 - 189

# Convert shares to integers
x_s = [share[0] for share in shares]
y_s = [int(share[1], 16) for share in shares]

# Evaluate the polynomial at x = 3 to get the third share
x_value = 3
third_share_value = lagrange_interpolation(x_value, x_s, y_s, p)
third_share_hex = format(third_share_value, 'x')

print("Constructed third share (x=3):", third_share_hex)
```
